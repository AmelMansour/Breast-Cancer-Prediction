```python
import pandas as pd
import re

def load_and_clean_data(input_file, output_file):
    # Charger le DataFrame nettoyé à partir du fichier d'entrée
    df = pd.read_csv(input_file)

    # Expression régulière pour extraire les informations entre parenthèses (p.Glu1527Lys)
    pattern = r'\(p\.([A-Za-z]+)(\d+)([A-Za-z]+)\)'

    # Appliquer l'expression régulière sur la colonne "Name" pour extraire les informations
    extracted_info = df['Name'].str.extract(pattern)

    # Ajouter les colonnes "AcideAminéSauvage", "Position" et "AcideAminéMuté" au DataFrame
    df['AcideAmineSauvage'] = extracted_info[0]
    df['Position'] = extracted_info[1]
    df['AcideAmineMute'] = extracted_info[2]

    # Sélectionner les colonnes à afficher
    columns_to_display = ['Name', 'AcideAmineSauvage', 'Position', 'AcideAmineMute', 'Clinical significance (Last reviewed)', 'GRCh37Location', 'GRCh38Location']
    displayed_df = df[columns_to_display]

    # Enregistrer le DataFrame affiché dans un fichier CSV
    displayed_df.to_csv(output_file, index=False)

    return displayed_df

input_file = 'BRCA.csv'
output_file = 'displayed_dataframe.csv'
displayed_df = load_and_clean_data(input_file, output_file)
display(displayed_df)
```

```python
import pandas as pd

# Charger le DataFrame à partir du fichier "displayed_dataframe.csv"
displayed_df = pd.read_csv('displayed_dataframe.csv')

def clean_displayed_data(df):
    # Supprimer les lignes dupliquées
    df = df.drop_duplicates()

    # Supprimer les lignes avec des données manquantes (NaN)
    df = df.dropna()

    # Réindexer le DataFrame après les suppressions
    df = df.reset_index(drop=True)

    # Supprimer les lignes contenant des valeurs "None" dans n'importe quelle colonne
    df = df.dropna(how='any')

    # Corriger les valeurs incorrectes dans la colonne "Position"
    def correct_position(position):
        try:
            # Tente de convertir la valeur de la position en un nombre entier
            position = int(position)
            return position
        except ValueError:
            # En cas d'erreur, renvoie NaN
            return None

    df['Position'] = df['Position'].apply(correct_position)

    # Supprimer les données manquantes (NaN) dans la colonne AcideAmineSauvage
    df = df.dropna(subset=['AcideAmineSauvage'])

    return df

# Nettoyer les données
cleaned_df = clean_displayed_data(displayed_df)

# Afficher le DataFrame nettoyé
display(cleaned_df)
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output
from datetime import datetime
from fpdf import FPDF
from google.colab import files

# Charger le DataFrame à partir du fichier "displayed_dataframe.csv"
displayed_df = pd.read_csv('displayed_dataframe.csv')

# Créer des listes d'acides aminés
acides_amines = ['Pro', 'Gln', 'Leu', 'Tyr', 'Glu', 'Ile', 'Val', 'Ser', 'Asp', 'Thr', 'Gly', 'Asn', 'His', 'Arg', 'Lys', 'Trp', 'Phe', 'Met', 'Ala', 'Cys']

# Liste pour stocker les résultats de filtrations
resultats_filtrations = []

# Initialiser le widget de sortie
output_text = widgets.Output()

# Variable pour garder trace de la dernière position filtrée
derniere_position_filtree = None
filtered_result = None  # Définir la variable en dehors de la fonction

class PDF(FPDF):
    def header(self):
        self.set_font("Arial", "B", 12)
        self.cell(0, 10, "Résultats de Filtration", align="C", ln=True)

    def chapter_title(self, title):
        self.set_font("Arial", "B", 12)
        self.cell(0, 10, title, ln=True)

    def chapter_body(self, body):
        self.set_font("Arial", size=12)
        self.multi_cell(0, 10, body)
        self.ln()

def generate_pdf(results, file_name):
    pdf = PDF()
    pdf.add_page()

    for i, resultat in enumerate(results, start=1):
        pdf.chapter_title(f"Résultat de la filtration {i} :")

        if not resultat.empty:  # Vérifiez si le DataFrame n'est pas vide
            resultat_text = "\n".join([f"{col} : {val}" for col, val in resultat.iloc[0].items()])
            pdf.chapter_body(resultat_text)
        else:
            pdf.chapter_body("Aucun résultat trouvé")

    pdf.output(file_name)

def filter_records(acide_amine_sauvage, position):
    resultat_filtre = displayed_df[
        (displayed_df['Position'] == position) &
        (
            displayed_df['AcideAmineSauvage'].str.contains(acide_amine_sauvage.replace('(', r'\('), case=False) |
            displayed_df['AcideAmineSauvage'].str.contains(acide_amine_sauvage.replace('(', r'\(').lower())
        )
    ]

    return resultat_filtre

def on_continue_button_click(b):
    output_text.clear_output()
    display_widgets()

def on_stop_button_click(b):
    global acide_amine_sauvage, position
    print("Exécution terminée.")
    current_date_time = datetime.now()
    print("             CHU FARHET HACHED Sousse ")
    print("            ------------------------------------")
    print("Date et heure actuelles:", current_date_time.strftime('%d/%m/%Y %H:%M:%S'))
    print("You have entered amino acid", acide_amine_sauvage, "and position", position)
    print("CHU FARHET HACHED Sousse")

    print("Résultats de toutes les filtrations effectuées:")
    for i, resultat in enumerate(resultats_filtrations):
        print(f"Résultat de la filtration {i + 1} :")
        if not resultat.empty:  # Vérifiez si le DataFrame n'est pas vide
            resultat_text = "\n".join([f"{col} : {val}" for col, val in resultat.iloc[0].items()])
            print(resultat_text)
        else:
            print("Aucun résultat trouvé")
        print()  # Ligne vide entre les résultats

    # Générez le PDF
    generate_pdf(resultats_filtrations, 'resultats_filtration.pdf')
    print("Le PDF des résultats a été généré avec succès : resultats_filtration.pdf")

    # Téléchargez le fichier PDF sur votre bureau
    files.download('resultats_filtration.pdf')

# Gestionnaire de lien
link = widgets.HTML(value="")
def on_link_clicked(change):
    if change.new == 'download_results':
        for i, resultat in enumerate(resultats_filtrations):
            fichier_resultat = f"resultat_filtration_{i + 1}.csv"
            resultat.to_csv(fichier_resultat, index=False)
        link.value = f"Résultats téléchargés : <a href='{fichier_resultat}' download>Télécharger les résultats</a>"
link.observe(on_link_clicked, 'value')

def display_results(acide_amine_sauvage, position):
    global derniere_position_filtree, filtered_result
    if position != derniere_position_filtree:  # Vérifier si la position a changé depuis la dernière filtration
        derniere_position_filtree = position
        filtered_result = filter_records(acide_amine_sauvage, position)
        resultats_filtrations.append(filtered_result)  # Ajouter le résultat de cette filtration à la liste

    with output_text:
        clear_output(wait=True)
        for i, resultat in enumerate(resultats_filtrations):
            print(f"Résultat de la filtration {i + 1} :")
            if not resultat.empty:  # Vérifiez si le DataFrame n'est pas vide
                resultat_text = "\n".join([f"{col} : {val}" for col, val in resultat.iloc[0].items()])
                print(resultat_text)
            else:
                print("Aucun résultat trouvé")
            print()  # Ligne vide entre les résultats

        continue_button = widgets.Button(description="Continuer")
        stop_button = widgets.Button(description="Arrêter")

        continue_button.on_click(on_continue_button_click)
        stop_button.on_click(on_stop_button_click)

        display(widgets.HBox([continue_button, stop_button]))

def display_widgets():
    acide_amine_dropdown = widgets.Dropdown(options=acides_amines, description="Type d'acide aminé:")
    position_entry = widgets.IntText(description="Position:")
    filter_button = widgets.Button(description="Filtrer")

    def handle_filter_button_click(b):
        global acide_amine_sauvage, position
        acide_amine_sauvage = acide_amine_dropdown.value
        position = position_entry.value
        display_results(acide_amine_sauvage, position)

    filter_button.on_click(handle_filter_button_click)

    display(acide_amine_dropdown, position_entry, filter_button, output_text)

# Afficher les widgets dans le bloc-notes
display_widgets()
display(link)
```
