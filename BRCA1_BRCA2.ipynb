```python
# Charger le premier fichier CSV dans un DataFrame
df1 = pd.read_csv('BRCA1.csv')
df1
```

```python
# Charger le deuxième fichier CSV dans un autre DataFrame
df2 = pd.read_csv('BRCA2.csv')
df2
```

```python
print(df1.head())
```

```python
print(df2.head())
```

```python
# Filtrer les gènes codants en supprimant les espaces autour de 'Protein coding'
df_codants1 = df1[df1['Biotype'].str.strip() == 'Protein coding']
df_codants1
```

```python
df_codants2 = df2[df2['Biotype'].str.strip() == 'Protein coding']
df_codants2
```

```python
# Combinez les deux DataFrames en un seul
combined_codants = pd.concat([df_codants1, df_codants2], ignore_index=True)
print(combined_codants.head())
combined_codants
```

```python
from google.colab import files

# Chemin local pour enregistrer le fichier CSV
csv_path_combined = '/content/genes_codants_combines.csv'

# Enregistrez le DataFrame combiné dans un fichier CSV local
combined_codants.to_csv(csv_path_combined, index=False)

# Téléchargez le fichier vers votre ordinateur
files.download(csv_path_combined)
```
